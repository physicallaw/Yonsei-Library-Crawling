# 연세대학교 도서관 장서 크롤링

## 사용방법
1. https://library.yonsei.ac.kr/ 에서 크롤링을 시작할 책을 검색한다.
2. 검색한 책의 상세페이지의 주소에서 CAT 이후의 12자리 숫자를 복사하여 `startnum`에 넣는다.
3. 크롤링을 종료할 책의 청구기호를 `endnum`에 넣는다.
4. 크롤링 결과를 저장할 파일 이름을 `filename`에 넣는다.

```python
import pandas as pd
import requests
from time import sleep
```

```python
startnum = "000001049543"
endnum = '025.1'
filename = 'computer'
```

```python
url = f"https://library.yonsei.ac.kr/search/callNoBrowsing/{startnum}"
params = {'type': 'next'}
response = requests.post(url)
df = pd.DataFrame(response.json())

while df.tail(1)['call_no'].values[0] < endnum:
    booknum = df.tail(1)['controlno'].values[0]
    url = f"https://library.yonsei.ac.kr/search/callNoBrowsing/{booknum}"
    response = requests.post(url, params)
    tempdf = pd.DataFrame(response.json())
    df = pd.concat([df, tempdf])
    print(df.tail(1)['call_no'].values[0] + " " + df.tail(1)['title'].values[0], end='\r')
    sleep(0.01)

df = df[['author','title','controlno','call_no']]
print('\n 크롤링 완료')
df.to_csv(f'{filename}.csv', index=False)
```

```python
startnum = "000000443428"
endnum = '016'
filename = 'computer'
```

```python
url = f"https://library.yonsei.ac.kr/search/callNoBrowsing/{startnum}"
params = {'type': 'pre'}
response = requests.post(url)
df2 = pd.DataFrame(response.json())

while df2.tail(5)['call_no'].values[0] > endnum:
    booknum = df2.tail(1)['controlno'].values[0]
    url = f"https://library.yonsei.ac.kr/search/callNoBrowsing/{booknum}"
    response = requests.post(url, params)
    tempdf = pd.DataFrame(response.json())
    df2 = pd.concat([df2, tempdf])
    print(df2.tail(1)['call_no'].values[0] + " " + df2.tail(1)['title'].values[0], end='\r')
    sleep(0.01)

df2 = df2[['author','title','controlno','call_no']]
print('\n 크롤링 완료')
df2.to_csv(f'{filename}.csv', index=False)
```
